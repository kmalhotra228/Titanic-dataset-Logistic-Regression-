```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression 
```

```python
# taining dataframe 
data = pd.read_csv('0000000000002429_training_titanic_x_y_train.csv')
data.head()
y_train = np.array(data['Survived'])
y_train
```

```python
data_train = data.copy()
data_train.head()
del data_train['Survived']
data_train.tail()

```

```python
def sex(string):
    if string == 'male':
        return 0
    else:
        return 1
```

```python
def Age(n):
    n = str(n)
    if n != 'nan':
        return int(float(n))
    else:
        return 0
    
```

```python
def change_dtype(dtype):
    dtype = str(dtype)
    if dtype != 'nan':
        return int(float(dtype))
    else:
        return 0
    
```

```python
def cabin(n):
    n = str(n)
    if n != 'nan':
        return 1
    else:
        return 0
    
```

```python
def embarked(n):
    n = str(n)
    if n == 'C':
        return 1
    elif n == 'Q':
        return 2
    elif n == 'S':
        return 3
    else:
        return 0
    
```

```python
data_train.head()
```

```python
new_dataframe = pd.DataFrame()
new_dataframe['Age'] = data_train['Age'].apply(Age)
new_dataframe['Sex'] = data_train['Sex'].apply(sex)
new_dataframe['SibSp'] = data_train['SibSp'].apply(change_dtype)
new_dataframe['Parch'] = data_train['Parch'].apply(change_dtype)
new_dataframe['Fare'] = data_train['Fare'].apply(change_dtype)
new_dataframe['Embarked'] = data_train['Embarked'].apply(embarked)
new_dataframe['Cabin'] = data_train['Cabin'].apply(cabin)
new_dataframe['Pclass'] = data_train['Pclass'].apply(change_dtype)
new_dataframe['Age_Age'] = new_dataframe['Age']*new_dataframe['Age']
new_dataframe['Age_embarked'] = new_dataframe['Age']*new_dataframe['Embarked']

new_dataframe.head()




```

```python
x_train = np.array(new_dataframe)
alg = LogisticRegression()
alg.fit(x_train,y_train)
```

```python
test_data = pd.read_csv('0000000000002429_test_titanic_x_test.csv')
test_data.head()
```

```python
test_dataframe = pd.DataFrame()
test_dataframe['Age'] = test_data['Age'].apply(Age)
test_dataframe['Sex'] = test_data['Sex'].apply(sex)
test_dataframe['SibSp'] = test_data['SibSp'].apply(change_dtype)
test_dataframe['Parch'] = test_data['Parch'].apply(change_dtype)
test_dataframe['Fare'] = test_data['Fare'].apply(change_dtype)
test_dataframe['Embarked'] = test_data['Embarked'].apply(embarked)
test_dataframe['Cabin'] = test_data['Cabin'].apply(cabin)
test_dataframe['Pclass'] = test_data['Pclass'].apply(change_dtype)
test_dataframe['Age_Age'] = test_dataframe['Age']*test_dataframe['Age']
test_dataframe['Age_embarked'] = test_dataframe['Age']*test_dataframe['Embarked']
test_dataframe.head()

```

```python
x_test = np.array(test_dataframe)
y_predict = alg.predict(x_test)
y_predict
```

```python
alg.score(x_train,y_train)
```

```python
alg.score(x_test,y_predict)
```
